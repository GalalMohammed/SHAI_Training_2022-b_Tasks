### In the following questions, use Pandas and Matplotlib to visualize Company Sales Data.
- Note: Read this data using Pandas or NumPy or using in-built matplotlib function.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/company_sales_data.csv")
df
```

#### 1. Using a line plot visualize the Total profit data provided for each month.

**Note:**It must be implemented in three methods
- Generated line plot must include the following properties:

X axis name = Month Number <br>
Y axis name = Total profit <br>
The line plot graph should look like this.


```python
####################################### Line Plot using Matplotlib visualization #########################################
plt.plot(df["month_number"], df["total_profit"])
plt.title("Company profit per month")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
```

```python
####################################### Line Plot using Seaborn visualization #########################################
sns.lineplot(x = df["month_number"], y = df["total_profit"])
plt.title("Company profit per month")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
```

```python
#################################### Line Plot using Pandas visualization #####################################
df.plot.line("month_number", "total_profit", title = "Company profit per month")
```

#### 2. Read all product sales data and display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).
**Note:**It must be implemented in three methods  
The graph should look like this.


```python
############################## multiline plots using Matplotlib visualization #########################################
for c in df.columns[1:-2]:
  plt.plot(df["month_number"], df[c], label = c + " Sales Data")
plt.legend()
plt.title("Sales Data")
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
```

```python
########################### multiline plots Plot using Seaborn visualization #########################################
for c in df.columns[1:-2]:
  sns.lineplot(x = df["month_number"], y = df[c], label = c + " Sales Data")
plt.legend()
plt.title("Sales Data")
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
```

```python
#################################### multiline plots using Pandas visualization #####################################
df.plot.line("month_number", df.columns[1:-2], title = "Sales Data")
```

#### 3. Read face cream and facewash product sales data and show it using the bar chart.


- The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.  
- **Note:**It must be implemented in three methods  

The bar chart should look like this.


```python
#################################### bar Plot using Matplotlib visualization #########################################
fig, ax = plt.subplots()
ax.bar(df["month_number"], df["facecream"], label = "facecream Sales Data", align = "edge")
ax.bar(df["month_number"], df["facewash"], label = "facewash Sales Data")
ax.set_title("Facewash and facecream Sales Data")
ax.set_xlabel("Month Number")
ax.set_ylabel("Sales Units in numbers")
plt.grid(True, linestyle = "--")
plt.legend()
```

```python
####################################### bar Plot using Seaborn visualization #########################################
dfCopy = df[["facecream", "facewash"]]
dfCopy.plot(kind = 'bar')
plt.title("Facewash and facecream Sales Data")
plt.xlabel("Month Number")
plt.ylabel("Sales Units in numbers")
plt.grid(True, linestyle = "--")
plt.legend()
```

```python
#################################### bar Plot using Pandas visualization #####################################
df.plot.bar("month_number", ["facecream", "facewash"], title = "Facecream and facewash Sales Data")
plt.grid(True, linestyle = "--")
```

#### 5. Read the total profit of each month and show it using the histogram to see the most common profit ranges.  
**Note:**It must be implemented in three methods.  
The histogram should look like this.


```python
####################################### histogram Plot using Matplotlib visualization #########################################
fig, ax = plt.subplots()
plt.hist(df['total_profit'])
ax.set_title("Profit Data")
ax.set_xlabel("Profit range in dollar")
ax.set_ylabel("Actual profit in dollar")
```

```python
####################################### histogram  Plot using Seaborn visualization #########################################
fig, ax = plt.subplots()
sns.histplot(df['total_profit'], ax = ax)
ax.set_title("Profit Data")
ax.set_xlabel("Profit range in dollar")
ax.set_ylabel("Actual profit in dollar")
```

```python
#################################### histogram Plot using Pandas visualization #####################################
fig, ax = plt.subplots()
ax = df['total_profit'].plot.hist()
ax.set_title("Profit Data")
ax.set_xlabel("Profit range in dollar")
ax.set_ylabel("Actual profit in dollar")
```

#### 6. Calculate total sale data for last year for each product and show it using a Pie chart.
- **Note:**
    - In Pie chart display Number of units sold per year for each product in percentage.
    - It must be implemented in two methods.

The Pie chart should look like this.


```python
#################################### pie Plot using Matplotlib visualization #####################################
fig, ax = plt.subplots()
ax.pie(df[df.columns[1:-2]].sum(), labels = df.columns[1:-2], autopct="%.1f%%")
ax.set_title('Sales Data')
```

```python
#################################### pie Plot using Pandas visualization #####################################
df[df.columns[1:-2]].sum().plot.pie(autopct="%.1f%%")
ax.set_title('Sales Data')
```
